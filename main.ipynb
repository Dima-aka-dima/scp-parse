```python
import requests
import time
import json
from bs4 import BeautifulSoup
```

```python
def download_pages():
    OBJECT_URL = 'https://the-scp.foundation/object'
    
    for i in range(2001, 4000):
        # make url
        n = '0'*(3-len(str(i))) + str(i)
        url = f"{OBJECT_URL}/scp-{n}"
        print(url)

        # get response
        r = requests.get(url)

        # write to a file
        file = open(f'./pages/scp-{n}.html', 'wb')
        file.write(r.content)
        file.close()
```

```python
download_pages()
```

```python
objects = {}

for i in range(1, 2000):
    
    n = '0'*(3-len(str(i))) + str(i)
    
    file = open(f'./pages/scp-{n}.html', 'rb')
    html = file.read()
    file.close()

    objects[n] = {}
    
    soup = BeautifulSoup(html, 'html.parser')
    
    objects[n]['name'] = soup.find("h2", {"class": "scp-nickname"}).text
    
    scp_tags = soup.find_all("div", {"class": "scp-tag"})
    if len(scp_tags) != 0:
        objects[n]['tags'] = []
        objects[n]['class'] = scp_tags[0].span.text
        for tag in scp_tags[1:]:
            objects[n]['tags'] += [tag.span.text]
    
    div_scp = soup.find('div', {'class': 'scp-special-containment-procedures'})
    if div_scp is not None:
        objects[n]['Special Containment Procedures'] = div_scp.text
    
    div_desc = soup.find('div', {'class': 'scp-description'})
    if div_scp is not None:
        objects[n]['Description'] = div_desc.text
        
```

```python
file = open('objects.json', 'w')
json.dump(objects, file, indent = 2)
```
